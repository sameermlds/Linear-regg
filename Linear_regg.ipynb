```python
data_one = pd.get_dummies(data=data)
```

```python
sns.scatterplot(x='horsepower',y='price',data=data);
```

```python
def normalize (x): 
    return ( (x-np.min(x))/ (max(x) - min(x)))
data_one_n = data_one.apply(normalize)
```

```python
pd.options.display.max_columns=None
```

```python
data_one_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_one_n.drop(labels=["price"],axis=1)
y = data_one_n[['price']]
```

```python
X.info()

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=700)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.hist()
```

```python
y_train.hist()
```

```python
#importing model and creating object 
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_predict = reg.predict(X_test)
```

```python
print("coeff\n",reg.coef_)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print("Model train parameters\n")
print("R2_Score\n",r2_score(y_test,y_predict))
print("mean_squared_error\n",mean_squared_error(y_test,y_predict))
```
